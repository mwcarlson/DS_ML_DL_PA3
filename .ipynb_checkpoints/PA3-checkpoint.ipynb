## Model Architecture / Results and Analysis
### My model architecture starts with a Sequential base model, which was chosen for its simplicity and ability to easily experiment with different types of layers. We could have used a Functional base model from keras, but we have no real need for complexities like branching/sharing layers and multiple inputs and outputs. My first layer with all of my models is an Input layer that is designed to take in the compressed, preprocessed form of each image. Next, I have a number of convolution layers (Conv2D) each followed by a pooling layer (MaxPooling2D), which applies filters to the input image to create a feature map and then uses pooling to reduce the spatial dimensions of the output for the next layer, essentially "zooming in" on our image. The "relu" activation function was used to allow our model to learn more complex patterns. The Flatten layer is used to convert the multidimensional input from the previous layers into a one-dimensional output that can be used by our following Dense layers. Our first Dense layer performs the classification steps we want, then the following Dropout layer drops a certain portion of our data randomly to prevent overfitting. Finally, our last Dense layer has only a single unit and is used to calculate a probability for each image using the sigmoid function. Our overall model is then compiled using the "adam" optimizer for faster convergence and a binary_crossentropy loss function because our results are expected to be binary.
### As far as hyperparameter and architecture tuning, I increased the number of filters by double and increased the kernel size for each convolution layer for my second model, as well as doubling the number of neurons in the first Dense layer. The idea was to sacrifice execution time for better results. For my third model, I kept all hyperparameters the same, but added an extra Conv2D and MaxPooling2D layer to test the effects.

```python
from keras.preprocessing import image_dataset_from_directory
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from keras.metrics import AUC
import tensorflow as tf
from PIL import Image
import time

res_model_num = [1, 2, 3]
res_loss = []
res_auc = []
res_time = []

train_labels["id"] = train_labels["id"].apply(lambda x: os.path.join(train_dir, x + ".tif"))

def read_tiff(image_path):
    image = Image.open(image_path.numpy())
    image = image.resize((96, 96))
    image_array = np.array(image)
    if image_array.ndim == 2:
        image_array = np.stack((image_array,) * 3, axis=-1)
    elif image_array.shape[2] == 1:
        image_array = np.squeeze(image_array, axis=2)
        image_array = np.stack((image_array,) * 3, axis=-1)
    return image_array

def preprocess_image(image_path, label):
    [image,] = tf.py_function(read_tiff, [image_path], [tf.uint8])
    image.set_shape([96, 96, 3])
    image = tf.cast(image, tf.float32) / 255.0
    label = tf.cast(label, tf.float32)
    return image, label

def load_dataset(df, batch_size):
    image_paths = df["id"].values
    labels = df["label"].values
    dataset = tf.data.Dataset.from_tensor_slices((image_paths, labels))
    dataset = dataset.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    dataset = dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
    return dataset

train_df = train_labels.sample(frac=0.8, random_state=64)
val_df = train_labels.drop(train_df.index)

batch_size = 32
train_dataset = load_dataset(train_df, batch_size)
val_dataset = load_dataset(val_df, batch_size)

time1 = time.time()

model = Sequential([
    Input((96, 96, 3)),
    Conv2D(32, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation="relu"),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=[AUC()])

epochs = 1
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs
)

test_image_paths = [os.path.join(test_dir, img) for img in os.listdir(test_dir) if img.endswith(".tif")]

test_dataset = tf.data.Dataset.from_tensor_slices(test_image_paths)
test_dataset = test_dataset.map(lambda image_path: preprocess_image(image_path, 0)[0])  # Labels are not used
test_dataset = test_dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

predictions = model.predict(test_dataset)

res_time.append((time.time() - time1) / 60)

predicted_labels = (predictions > 0.5).astype(int)

submission_df = pd.DataFrame({
    "id": [os.path.basename(path)[:-4] for path in test_image_paths],
    "label": predicted_labels.flatten(),
})

submission_df.to_csv("submission1.csv", index=False)

loss, auc = model.evaluate(val_dataset)
res_loss.append(loss)
res_auc.append(auc)

print(f"\nModel 1 Summary:\n{model.summary()}")

# Model 2

time1 = time.time()

model = Sequential([
    Input((96, 96, 3)),
    Conv2D(64, (5, 5), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(128, (5, 5), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(256, (5, 5), activation="relu"),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(1028, activation="relu"),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=[AUC()])

history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs
)

predictions = model.predict(test_dataset)

res_time.append((time.time() - time1) / 60)

predicted_labels = (predictions > 0.5).astype(int)

submission_df = pd.DataFrame({
    "id": [os.path.basename(path)[:-4] for path in test_image_paths],
    "label": predicted_labels.flatten(),
})

submission_df.to_csv("submission2.csv", index=False)

loss, auc = model.evaluate(val_dataset)
res_loss.append(loss)
res_auc.append(auc)

print(f"\nModel 2 Summary:\n{model.summary()}")

# Model 3

time1 = time.time()

model = Sequential([
    Input((96, 96, 3)),
    Conv2D(32, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(256, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation="relu"),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=[AUC()])

history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs
)

predictions = model.predict(test_dataset)

res_time.append((time.time() - time1) / 60)

predicted_labels = (predictions > 0.5).astype(int)

submission_df = pd.DataFrame({
    "id": [os.path.basename(path)[:-4] for path in test_image_paths],
    "label": predicted_labels.flatten(),
})

submission_df.to_csv("submission3.csv", index=False)

loss, auc = model.evaluate(val_dataset)
res_loss.append(loss)
res_auc.append(auc)

print(f"\nModel 3 Summary:\n{model.summary()}")
```

Model 1 had about half the execution time and number of parameters compared to model 2 but ended up with slightly better results. Finally, model 3 had the best results, with the fastest execution times and best scores. The execution time likely happened because most of model 1 was still fresh in memory, and model 3 was the same as model 1 but with an extra convolution and pooling step, so it ended up with the best loss and AUC scores due to the refinement of those extra steps.
