# Week 3: CNN Cancer Detection Kaggle Mini-Project

## Description
### This week's project is based on a Kaggle competition in which our task is to identify metastatic cancer in a series of small images that are patches taken from larger digital pathology scans. If an image is identified as showing cancer, it will receive a 1 (true) label, and a 0 (false) label otherwise.
### The data itself is supplied in different forms. In order to train with our training data, we are supplied with a CSV file containing an "id" column containing the ID for each image in the training set as well as a corresponding label in the "label" column for whether cancer was detected in each image or not. This file contains 220025 rows and 2 columns. Additionally, the data download contains subfolders comprised of all of the training and test images, with 220025 files in the "train" subfolder and 57458 files in the "test" subfolder. The file name of each image (minus the file extension) corresponds to the "id" column of the labeled CSV file.

```python
import pandas as pd
import numpy as np
import os

base_dir = "C:\\Users\\mcarlson\\Downloads\\histopathologic-cancer-detection\\"
train_labels = pd.read_csv(base_dir + "train_labels.csv")
print(f"Shape of training labels dataframe: {train_labels.shape}")
print(f"\nHead of dataframe:\n{train_labels.head()}")
train_dir = base_dir + "train\\"
test_dir = base_dir + "test\\"
train_file_names = [name for name in os.listdir(train_dir) if os.path.isfile(os.path.join(train_dir, name))]
print(f"\nNumber of image files in training data: {len(train_file_names)}")
test_file_names = [name for name in os.listdir(test_dir) if os.path.isfile(os.path.join(test_dir, name))]
print(f"Number of image files in test data: {len(test_file_names)}")
train_ids = [os.path.splitext(name)[0] for name in train_file_names]
test_ids = [os.path.splitext(name)[0] for name in test_file_names]
```

## Model Architecture
### t

```python
from keras.preprocessing import image_dataset_from_directory
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.metrics import AUC
import tensorflow as tf
from PIL import Image

train_labels["id"] = train_labels["id"].apply(lambda x: os.path.join(train_dir, x + ".tif"))
#train_labels["label"] = train_labels["label"].astype(str)

def read_tiff(image_path):
    # Open the image file with PIL (which supports .tif)
    image = Image.open(image_path.numpy())
    image = image.resize((96, 96))
    image_array = np.array(image)
    # Convert single-channel images to three channels
    if image_array.ndim == 2:
        image_array = np.stack((image_array,) * 3, axis=-1)
    # Ensure image has three channels (for RGB)
    elif image_array.shape[2] == 1:
        image_array = np.squeeze(image_array, axis=2)
        image_array = np.stack((image_array,) * 3, axis=-1)
    return image_array

# Define function for preprocessing images
def preprocess_image(image_path, label):
    [image,] = tf.py_function(read_tiff, [image_path], [tf.uint8])
    image.set_shape([96, 96, 3])
    image = tf.cast(image, tf.float32) / 255.0
    label = tf.cast(label, tf.float32)
    return image, label

# Function to load dataset from file paths and labels
def load_dataset(df, batch_size):
    image_paths = df["id"].values
    labels = df["label"].values
    dataset = tf.data.Dataset.from_tensor_slices((image_paths, labels))
    dataset = dataset.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    dataset = dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
    return dataset

# Prepare the training and validation datasets
train_df = train_labels.sample(frac=0.8, random_state=64)  # 80% training data
val_df = train_labels.drop(train_df.index)  # 20% validation data

# Load the datasets
batch_size = 32
train_dataset = load_dataset(train_df, batch_size)
val_dataset = load_dataset(val_df, batch_size)

model = Sequential([
    Conv2D(32, (3, 3), activation="relu", input_shape=(96, 96, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation="relu"),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=[AUC()])

epochs = 15
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs
)

test_image_paths = [os.path.join(test_dir, img) for img in os.listdir(test_dir) if img.endswith(".tif")]

# Create a test dataset
test_dataset = tf.data.Dataset.from_tensor_slices(test_image_paths)
test_dataset = test_dataset.map(lambda image_path: preprocess_image(image_path, 0)[0])  # Labels are not used
test_dataset = test_dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

predictions = model.predict(test_dataset)

# Assuming the Kaggle submission requires a binary classification output
predicted_labels = (predictions > 0.5).astype(int)

# Prepare the submission DataFrame
submission_df = pd.DataFrame({
    "id": [os.path.basename(path)[:-4] for path in test_image_paths],
    "label": predicted_labels.flatten(),
})

submission_df.to_csv("submission.csv", index=False)
```
